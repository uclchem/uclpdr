```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Square Wave

```python
df=pd.read_csv("cloud-1d-n=1e3.dat")
```

```python
len(df)
density=10000
for i in range(15):
    df.iloc[i*100:,4]=density
    if density>110:
        density=1000
    else:
        density=100000
```

```python
fig,ax=plt.subplots()
df.plot(y="  n_H (c^m-3)  ",x="Particle # ",ax=ax)
```

```python
df.to_csv("cloud-square.dat",index=False)
```

# Rising Density

```python
df=pd.read_csv("cloud-1d-n=1e3.dat")
```

```python
n_points=len(df)
density=np.linspace(100,1e5,n_points)
```

```python
df["  n_H (c^m-3)  "]=density
```

```python
fig,ax=plt.subplots()
df.plot(y="  n_H (c^m-3)  ",x="Particle # ",ax=ax)
```

```python
df.to_csv("linear-density-increase.dat",index=False)
```

# Falling Density

```python
df=pd.read_csv("cloud-1d-n=1e3.dat")
n_points=len(df)
density=np.linspace(1e5,100,n_points)
df["  n_H (c^m-3)  "]=density
```

```python
fig,ax=plt.subplots()
df.plot(y="  n_H (c^m-3)  ",x="Particle # ",ax=ax)
```

```python
df.to_csv("linear-density-decrease.dat",index=False)
```
